```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
father=[[65],[63.3],[65],[65.8],[61.1]]
son=[[59.8],[63.2],[63.3],[62.8],[64.3]]
father=np.array(father)
son=np.array(son)

x_train=father[0:3]
y_train=son[0:3]
x_test=father[3:]
y_test=son[3:]

lm=LinearRegression()
lm.fit(x_train,y_train)
predictions=lm.predict(x_test)
print(predictions)

plt.scatter(x_train,y_train,color="b")
plt.plot(x_test,predictions,color='black',linewidth=3)
plt.title("Fathet V/s Son",color='g')
plt.xlabel('Father')
plt.ylabel('Son')
plt.show()
```
